```python
from jupyter_dash import JupyterDash
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic


```

```python
df = pd.read_csv("IPL2019.csv")
df.head(15)
```

```python
# Initialize the app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    
    html.H1("IPL Statistics Report for 2019 "),
    
    # Histogram
    dcc.Graph(
        id='histogram',
        figure={
            'data': [{
                'x': df['Econ'],
                'type': 'histogram'
            }],
            'layout': {
                'title': 'Histogram',
                 'xaxis' : {'title' : 'Econ of a bowler'}
               
            }
            }
        
    ),
    
    # Box plot
    dcc.Graph(
        id='boxplot',
        figure={
            'data': [{
                'x': df['Player'],
                'y': df['Inns'],
                'type': 'box'
            }],
            'layout': {
                'title': 'Box plot',
                'xaxis' : {'title' : 'Player'},
                'yaxis' : {'title' : 'Innings plays'}
                
            }
        }
    ),
    # Mosaic plot
    dcc.Graph(
        id='mosaicplot',
        figure={
            'data': [{
                'y': df['SR'],
                'type': 'mosaicplot'
            }],
            'layout': {
                'title': 'Mosaic plot',
                'yaxis' : {'title' : 'straik rate'}
            }
        }
    ),
    
    # Scatter plot
    dcc.Graph(
        id='scatterplot',
        figure={
            'data': [{
                'x': df['Econ'],
                'y': df['SR'],
                'mode': 'Growth',
                'marker': {
                    'size': 15,
                    'color': 'orange',
                    'opacity': 0.7
                }
            }],
            'layout': {
                'title': 'Scatter plot',
                'xaxis' : {'title' : 'Econ'},
                'yaxis' : {'title' : 'straik rate'}
                
            }
        }
    ),
    
    # Bar chart
    dcc.Graph(
        id='barchart',
        figure={
            'data': [{
                'x': df['Player'],
                'y': df['Runs'],
                'type': 'bar'
            }],
            'layout': {
                'title': 'Bar chart',
                'xaxis' : {'title' : 'Player'},
                'yaxis' : {'title' : 'Runs'}
                
                
            }
        }
    )
])


```

```python
if __name__ == '__main__':
    app.run_server(port=1050)
```
